```python
import sympy as sp
```

```python
x = sp.Symbol('x')
expr = sp.sin(x)**2 - sp.cos(x)**2
expr
```

```python
expr_solved = sp.solveset(expr)
expr_solved
```

```python
graph = sp.plot(expr)
graph
```

```python
expr_diff = sp.diff(expr)
expr_diff
```

```python
rise_expr = sp.solveset(expr_diff>0)
rise_expr
```

```python
down_expr = sp.solve(expr_diff<0)
down_expr
```

```python
top_expr = sp.solveset(expr_diff)
top_expr
```

```python
rise_expr_origin = sp.solveset(expr>0)
rise_expr_origin
```

```python
down_expr_origin = sp.solveset(expr<0)
down_expr_origin
```
